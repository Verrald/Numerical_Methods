```python
"""
================================================
  PENYELESAIAN SISTEM PERSAMAAN LINEAR  Ax = b

  Dua metode:
    1. Gaussian Elimination
    2. Gauss-Jordan Elimination
================================================
"""

import copy


# ── CETAK MATRIKS AUGMENTED [A|B] ─────────────────────────────────────────
def cetak_matriks(A, B, judul=""):
    n = len(A)
    if judul:
        print(f"\n  [{judul}]")
    print("  " + "-" * 40)
    for i in range(n):
        baris = "  | "
        for j in range(n):
            baris += f"{A[i][j]:8.4f} "
        baris += f"| {B[i]:8.4f} |"
        print(baris)
    print("  " + "-" * 40)


#  METODE 1: GAUSSIAN ELIMINATION

def gaussian_elimination(A_input, B_input):
    """
    Langkah:
      1. Forward Elimination  -> bentuk Upper Triangular
      2. Backward Substitution -> hitung x dari bawah ke atas
    """
    n = len(A_input)

    # salin supaya input asli tidak berubah
    A = copy.deepcopy(A_input)
    B = copy.deepcopy(B_input)

    print("\n" + "="*50)
    print("  METODE 1 : GAUSSIAN ELIMINATION")
    print("="*50)
    cetak_matriks(A, B, "Matriks Awal")

    # ── LANGKAH 1: FORWARD ELIMINATION ──────────
    print("\n  -- FORWARD ELIMINATION --")

    for k in range(n):          # k = baris pivot (0, 1, 2, ...)

        # cek pivot = 0 (tidak bisa dibagi)
        if A[k][k] == 0:
            print(f"  ! Pivot baris {k+1} = 0, tidak bisa lanjut!")
            return None

        print(f"\n  Pivot ke-{k+1}: A[{k+1}][{k+1}] = {A[k][k]:.4f}")

        for i in range(k+1, n): # i = baris yang akan dieliminasi

            faktor = A[i][k] / A[k][k]
            print(f"  R{i+1} = R{i+1} - ({faktor:.4f}) x R{k+1}")

            for j in range(n):
                A[i][j] = A[i][j] - faktor * A[k][j]
            B[i] = B[i] - faktor * B[k]

        cetak_matriks(A, B, f"Setelah eliminasi kolom {k+1}")

    # ── LANGKAH 2: BACKWARD SUBSTITUTION ────────
    print("\n  -- BACKWARD SUBSTITUTION --")

    X = [0.0] * n

    for i in range(n-1, -1, -1):   # mulai dari baris paling bawah

        jumlah = 0
        for j in range(i+1, n):
            jumlah += A[i][j] * X[j]

        X[i] = (B[i] - jumlah) / A[i][i]
        print(f"  x{i+1} = ({B[i]:.4f} - {jumlah:.4f}) / {A[i][i]:.4f} = {X[i]:.6f}")

    return X


#  METODE 2: GAUSS-JORDAN ELIMINATION

def gauss_jordan(A_input, B_input):
    """
    Langkah:
      - Untuk setiap baris pivot:
          1. Bagi baris pivot agar A[k][k] = 1
          2. Nolkan semua baris LAIN di kolom k (atas & bawah)
      - Hasil akhir: A menjadi Identity Matrix
      - B langsung = solusi X (tidak perlu back-substitution)
    """
    n = len(A_input)

    A = copy.deepcopy(A_input)
    B = copy.deepcopy(B_input)

    print("\n" + "="*50)
    print("  METODE 2 : GAUSS-JORDAN ELIMINATION")
    print("="*50)
    cetak_matriks(A, B, "Matriks Awal")

    for k in range(n):          # k = kolom pivot

        if A[k][k] == 0:
            print(f"  ! Pivot baris {k+1} = 0, tidak bisa lanjut!")
            return None

        pivot = A[k][k]
        print(f"\n  Langkah {k+1}: normalisasi R{k+1} (bagi {pivot:.4f})")

        # bagi seluruh baris k agar A[k][k] = 1
        for j in range(n):
            A[k][j] = A[k][j] / pivot
        B[k] = B[k] / pivot

        # nolkan kolom k di semua baris lain
        for i in range(n):
            if i == k:
                continue
            if A[i][k] == 0:
                continue

            faktor = A[i][k]
            print(f"  R{i+1} = R{i+1} - ({faktor:.4f}) x R{k+1}")

            for j in range(n):
                A[i][j] = A[i][j] - faktor * A[k][j]
            B[i] = B[i] - faktor * B[k]

        cetak_matriks(A, B, f"Setelah langkah {k+1}")

    # B sekarang langsung = solusi
    X = list(B)
    return X


#  CETAK SOLUSI & CEK RESIDUAL
def tampilkan_hasil(X, A, B, nama_metode):
    if X is None:
        print("  Tidak ada solusi.\n")
        return

    n = len(X)
    print(f"\n  >> Solusi ({nama_metode}):")
    for i in range(n):
        print(f"     x{i+1} = {X[i]:.6f}")

#  CONTOH SOAL

if __name__ == "__main__":

    # Sistem persamaan 3x3:
    #
    #   2x1 +  x2 +  x3 =  5
    #    x1 + 3x2 +  x3 =  9
    #    x1 +  x2 + 4x3 =  7
    #
    # Solusi: x1 = 0.8235, x2 = 2.4118, x3 = 0.9412

    A = [
        [2, 1, 1],
        [1, 3, 1],
        [1, 1, 4]
    ]
    B = [5.0, 9.0, 7.0]

    print("\n  Sistem persamaan yang diselesaikan:")
    print("    2x1 +  x2 +  x3 = 5")
    print("     x1 + 3x2 +  x3 = 9")
    print("     x1 +  x2 + 4x3 = 7")

    # ── jalankan Gaussian Elimination
    X1 = gaussian_elimination(A, B)
    tampilkan_hasil(X1, A, B, "Gaussian Elimination")

    # ── jalankan Gauss-Jordan
    X2 = gauss_jordan(A, B)
    tampilkan_hasil(X2, A, B, "Gauss-Jordan")
```
