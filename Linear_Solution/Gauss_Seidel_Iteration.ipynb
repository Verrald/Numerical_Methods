```python
"""
================================================
  PENYELESAIAN SISTEM PERSAMAAN LINEAR  Ax = b

  Metode: Gauss-Seidel Iteration
================================================
"""


# ── CETAK HEADER TABEL ITERASI ────────────────────────────────────────────
def cetak_header(n):
    header = f"  {'Iterasi':<8}"
    for i in range(n):
        header += f"  {'x'+str(i+1):<14}"
    header += f"  {'Error':<14}"
    print(header)
    print("  " + "-" * (10 + 16*n + 16))


# ── CETAK BARIS ITERASI ───────────────────────────────────────────────────
def cetak_baris(iterasi, X, error):
    baris = f"  {iterasi:<8}"
    for xi in X:
        baris += f"  {xi:<14.8f}"
    baris += f"  {error:<14.2e}"
    print(baris)


# ── GAUSS-SEIDEL ITERATION ────────────────────────────────────────────────
def gauss_seidel(A, B, tol=1e-6, max_iter=100, mode='tolerance'):
    """
    Metode iteratif Gauss-Seidel untuk Ax = b.

    Prinsip:
      - Setiap nilai x[i] diperbarui LANGSUNG pada iterasi yang sama
        (menggunakan nilai x terbaru yang sudah dihitung)
      - Konvergensi dijamin jika A diagonal dominan

    Rumus:
      x[i] = (b[i] - sum(A[i][j]*x[j], j != i)) / A[i][i]

    Parameters
    ----------
    A        : list[list[float]]  -- matriks koefisien n x n
    B        : list[float]        -- vektor konstanta
    tol      : float              -- toleransi error (mode='tolerance')
    max_iter : int                -- jumlah iterasi maksimum (mode='iteration')
    mode     : str                -- 'tolerance' atau 'iteration'

    Returns
    -------
    X       : list[float] -- vektor solusi
    iterasi : int         -- jumlah iterasi yang dilakukan
    """
    n = len(A)

    # Cek diagonal dominan
    for i in range(n):
        diag = abs(A[i][i])
        jumlah_off = sum(abs(A[i][j]) for j in range(n) if j != i)
        if diag < jumlah_off:
            print(f"  ! Peringatan: baris {i+1} tidak diagonal dominan.")
            print("    Konvergensi tidak dijamin.\n")

    print("\n" + "="*50)
    print("  GAUSS-SEIDEL ITERATION")
    print("="*50)
    print(f"  Mode     : {mode}")
    if mode == 'tolerance':
        print(f"  Toleransi: {tol:.2e}")
    else:
        print(f"  Max iter : {max_iter}")
    print()

    # Tebakan awal: semua nol
    X = [0.0] * n
    iterasi = 0

    cetak_header(n)

    while True:
        iterasi += 1
        X_lama = X[:]

        for i in range(n):
            sigma = 0.0
            for j in range(n):
                if j != i:
                    sigma += A[i][j] * X[j]     # pakai X terbaru (sudah diupdate)
            X[i] = (B[i] - sigma) / A[i][i]

        # Error = maksimum perubahan absolut
        error = max(abs(X[i] - X_lama[i]) for i in range(n))

        cetak_baris(iterasi, X, error)

        if mode == 'tolerance' and error < tol:
            break
        elif mode == 'iteration' and iterasi >= max_iter:
            break

    return X, iterasi


# ── TAMPILKAN HASIL ───────────────────────────────────────────────────────
def tampilkan_hasil(X, iterasi, nama_metode):
    if X is None:
        print("  Tidak ada solusi.\n")
        return

    n = len(X)
    print(f"\n  >> Solusi ({nama_metode}):")
    for i in range(n):
        print(f"     x{i+1} = {X[i]:.6f}")
    print(f"     Iterasi: {iterasi}")


# ── CONTOH SOAL ───────────────────────────────────────────────────────────
if __name__ == "__main__":

    # Sistem persamaan 3x3 (diagonal dominan):
    #
    #   4x1 +  x2 +  x3 =  9
    #    x1 + 5x2 +  x3 = 10
    #    x1 +  x2 + 6x3 = 12
    #
    # Solusi: x1 = 1.3846, x2 = 1.2308, x3 = 1.6154

    A = [
        [4, 1, 1],
        [1, 5, 1],
        [1, 1, 6]
    ]
    B = [9.0, 10.0, 12.0]

    print("\n  Sistem persamaan yang diselesaikan:")
    print("    4x1 +  x2 +  x3 =  9")
    print("     x1 + 5x2 +  x3 = 10")
    print("     x1 +  x2 + 6x3 = 12")

    # Mode 1: Berdasarkan toleransi
    print("\n" + "─"*50)
    print("  MODE 1 - Toleransi")
    print("─"*50)
    X1, iter1 = gauss_seidel(A, B, tol=1e-6, mode='tolerance')
    tampilkan_hasil(X1, iter1, "Gauss-Seidel (Toleransi)")

    # Mode 2: Berdasarkan jumlah iterasi
    print("\n" + "─"*50)
    print("  MODE 2 - Iterasi Tetap")
    print("─"*50)
    X2, iter2 = gauss_seidel(A, B, max_iter=10, mode='iteration')
    tampilkan_hasil(X2, iter2, "Gauss-Seidel (Iterasi Tetap)")
```
